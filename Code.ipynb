HANDLING Missing Values

```python
match_df.isna().sum()
```

```python
match_df[match_df['city'].isnull()]
```

```python
venue_city_mapping = match_df[match_df['city'].notnull()].drop_duplicates(subset='venue').set_index('venue')['city'].to_dict()
```

```python
match_df['city'] = match_df.apply(lambda x : venue_city_mapping.get(x['venue'],x['city']) if pd.isnull(x['city']) else x['city'],
                                 axis = 1)
```

```python
match_df[['target_overs','method']].sort_values(by='target_overs').head(100)
```

```python
match_df['method'] = match_df['target_overs'].apply(lambda x : 'D/L' if pd.notnull(x) and  x < 20 else 'Normal')
```

```python
inconsistent_ties = match_df[(match_df['result'] == 'tie') & (match_df['winner'].notnull())]

inconsistent_ties[['result', 'winner', 'super_over']]
```

```python
match_df['winner'] = match_df['winner'].fillna('Draw')
```

```python
match_df.loc[match_df['result'].isin(['tie', 'no result']), 'result_margin'] = 0
```

```python
match_df['target_runs'] = match_df['target_runs'].fillna(0)
match_df['target_overs'] = match_df['target_overs'].fillna(0)
match_df['player_of_match'] = match_df['player_of_match'].fillna(0)
```

DELIVERIES DATAFRAME CLEANING

```python
deliveries_df.head()
```

```python
deliveries_df[['batting_team','bowling_team']] = deliveries_df[['batting_team','bowling_team']]\
                    .replace({'Royal Challengers Bangalore' : 'Royal Challengers Bengaluru' ,
                            'Rising Pune Supergiants' : 'Rising Pune Supergiant',
                            'Kings XI Punjab' : 'Punjab Kings' ,
                             'Delhi Daredevils' : 'Delhi Capitals'})
```

```python
deliveries_df.isna().sum()
```

```python
deliveries_df.groupby(['extras_type', 'extra_runs'], dropna=False).size().reset_index(name='count')
```

```python
deliveries_df['extras_type'] = deliveries_df['extras_type'].fillna('Legal')
```

```python
deliveries_df['dismissal_kind'] = deliveries_df['dismissal_kind'].fillna('Nil')
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].fillna('Nil')
deliveries_df['fielder'] = deliveries_df['fielder'].fillna('Nil')
```
